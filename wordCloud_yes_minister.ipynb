```python
filename = "yes-minister.txt"

with open(filename) as f:
    mytext = f.read()
```

```python
# mytext
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white",width=1000,height=880).generate(mytext)
```

```python
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.savefig('yes_minister_wordcloud2.pdf')
plt.show()
```
